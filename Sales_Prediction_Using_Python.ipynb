```python
import pandas as pd
import numpy as np
```

```python
#Reading CSV File
sales = pd.read_csv("/content/Advertising.csv")

sales.head() #print first 5 entries
```

```python
#dropping column

sales = sales.drop(columns = ["Unnamed: 0"])
```

```python
sales # return dataframe
```

```python
X = sales.iloc[:,0:-1]
X
```

```python
Y = sales.iloc[:,-1] #target
Y
```

```python
# Train Test split
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 43)
```

```python
X_train
```

```python
X_test
```

```python
X_train = X_train.astype(int)
Y_train = Y_train.astype(int)
X_test = X_test.astype(int)
Y_test = Y_test.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()

X_train_scaled = SC.fit_transform(X_train)
X_test_scaled = SC.fit_transform(X_test)
```

```python
# APPLY LINEAR REGRESSION

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train_scaled , Y_train)
```

```python
prediction1 = lr.predict(X_test_scaled)
```

```python
#EVALUATION THE PERFORMANCE

from sklearn.metrics import r2_score

r2_score(Y_test, prediction1)
```
